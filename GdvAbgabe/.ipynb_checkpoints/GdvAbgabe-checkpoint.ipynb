```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# LE1

```python
# Daten einlesen
dataFile1 = pd.read_csv("out_1.csv")
dataFile2 = pd.read_csv("out_2.csv")
#Zone hinzufügen
dataFile1["zone"]="1"
dataFile2["zone"]="2"
# Zonen zusammenfügen
steinschlaege = dataFile1.append(dataFile2)
steinschlaege.head()
```

```python
# Abbildung 1: Geschwindigkeit in m/s je nach Masse für Zone 1
plt.scatter(dataFile1["mass"],dataFile1["velocity"],color = "hotpink")
plt.axis([-100,3000,0,50])
plt.title("Geschwindigkeit in m/s je nach Masse für Zone 1")
plt.xlabel("Masse")
plt.ylabel("Gescheindigkeit in m/s")
plt.show()
```

```python
# Abbildung 2: Geschwindigkeit in m/s je nach Masse für Zone 2
plt.scatter(dataFile2["mass"],dataFile2["velocity"],color="hotpink")
plt.axis([-100,3000,0,50])
plt.title("Geschwindigkeit in m/s je nach Masse für Zone 2")
plt.xlabel("Masse")
plt.ylabel("Gescheindigkeit in m/s")
plt.show()
```

```python
# Abbildung 3: Geschwindigkeit in m/s nach Ablösungszone
x1= dataFile1["mass"]
x2=dataFile2["mass"]
y1=dataFile1["velocity"]
y2=dataFile2["velocity"]

plt.scatter(x1,y1,c="hotpink")
plt.scatter(x2,y2,c="darkgreen")
plt.xlabel("Masse")
plt.ylabel("Geschwindigkeit in m/s")
plt.title("Geschwindigkeit in m/s je nach Masse für Zone kombiniert / rosa: Zone 1, grün: Zone 2")
plt.show()
```

```python
# Abbildung 4 : Geschwindigkeit in m/s nach Ablösungszone 
steinschlaege.boxplot(column="velocity", by="zone",color="hotpink")
plt.title("Geschwindigkeit in m/s nach Ablösungszone")
plt.suptitle(" ")
plt.xlabel("Ablösungszone")
plt.ylabel("Geschwindigkeit in m/s")
plt.show()
```

# LE 2

```python
# Daten einlesen
# quelle : https://opendata.swiss/en/dataset/covid-19-schweiz/resource/ce8c184f-e1fb-4adf-b0a2-c1bc9ec8471a
data2 = pd.read_csv("PopulationAgeRangeSexData.csv")
data2.head()
```

```python
# Datentyp anpassen
data2.dtypes
```

```python
data2["geoRegion"] = data2["geoRegion"].astype("string")
data2.dtypes
```

```python
df = data2[(data2["geoRegion"]=="AG") | (data2["geoRegion"] =="ZH") |(data2["geoRegion"]=="SO")
          | (data2["geoRegion"] == "LU")]
#df.head()
```

```python
df_ag = df[df["geoRegion"] == "AG"]
df_zh = df[df["geoRegion"] == "ZH"]
df_so = df[df["geoRegion"] == "SO"]
df_lu = df[df["geoRegion"] == "LU"]
```

```python
# Gesamt Population pro Kanton por Altersklasse
ag = df_ag.groupby("ageRange").pop.sum()
#print("AG", ag)

zh = df_zh.groupby("ageRange").pop.sum()
#print("ZH", zh)

so = df_so.groupby("ageRange").pop.sum()
#print("SO", so)

lu = df_lu.groupby("ageRange").pop.sum()
#print("LU", lu)
```

```python
# Abbildung 7: Grafik vor Anpassung

altersklasse = ["0-9", "10-19", "20-29", "30-39", "40-49", "50-59","60-69","70-79","80+"]

ag_age = altersklasse
ag_pop = [ag][0]
plt.plot(ag_age, ag_pop,color = "black")

zh_age = altersklasse
zh_pop = [zh][0]
plt.plot(zh_age, zh_pop,color="black")

so_age = altersklasse
so_pop = [so][0]
plt.plot(so_age, so_pop,color="black")

lu_age = altersklasse
lu_pop = [lu][0]
plt.plot(lu_age, lu_pop,color="black")


legend = plt.legend(['AG','ZH',"SO","LU"], title = "Legende")
plt.xlabel('Altersklasse')
plt.ylabel('Bevölkerung')
plt.show()
```

```python
# Daten einlesen
diabetes = pd.read_excel("diabetes.xlsx",sheet_name=1,skiprows = 2)
diabetes.tail(20)
```

```python
# BMI Werte untersuchen
```

```python
# 0 Werte als NaN speichern und erneut Histogramm ausgeben
bmi_mit_nan = diabetes["bmi"].replace(0, np.NaN)
```

```python
# BMI Werte vor Datawrangling
plt.hist(diabetes["bmi"],bins = 20)
plt.axis([-2,80,0,150])
plt.axvline(diabetes["bmi"].median(), color='r', linestyle='dashed', linewidth=1)
plt.show()

# Es fällt auf, dass es einige 0 Werte hat. Dies ist unmöglich.
```

```python
plt.hist(bmi_mit_nan,bins = 15)
plt.axvline(bmi_mit_nan.median(), color='r', linestyle='dashed', linewidth=1)
plt.axis([-2,80,0,150])
plt.show()
```

### Frage: Dürfen Bins beliebig angepasst werden?

```python
# zusammen in einer Visualisierung
plt.hist(diabetes["bmi"],bins = 20,color="r")
plt.hist(bmi_mit_nan,bins = 20,color="b")
plt.axis([-2,80,0,150])
```

```python
print(bmi_mit_nan.mean())
print(diabetes["bmi"].mean())
```

```python
print(bmi_mit_nan.median())
print(diabetes["bmi"].median())
```
