```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Daten einlesen
dataFile1 = pd.read_csv("out_1.csv")
dataFile2 = pd.read_csv("out_2.csv")
#Zone hinzufügen
dataFile1["zone"]="1"
dataFile2["zone"]="2"
# Zonen zusammenfügen
steinschlaege = dataFile1.append(dataFile2)
steinschlaege
```

```python
dataFile2
```

```python
# Barchart pro Ablsöungszone, wann die Steine gefallen sind


steinschlaege[steinschlaege["mass"]<50]

```

```python
steinschlaege.boxplot(column="velocity", by="zone")
plt.title("Geschwindigkeit in m/s nach Ablösungszone")
plt.suptitle(" ")
plt.xlabel("Ablösungszone")
plt.ylabel("Geschwindigkeit in m/s")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(steinschlaege["mass"],steinschlaege["velocity"])
ax.set_title("Geschwindigkeit je nach Masse für beide Zonen")
ax.set_xlabel("Masse")
ax.set_ylabel("Geschwindigkeit")
```

```python
# Geschwindigkeit und Masse im Vergleich für Zone 1
fig, ax = plt.subplots()
ax.scatter(dataFile1["mass"],dataFile1["velocity"])
ax.set_title("Geschwindigkeit je nach Masse für Zone 1")
ax.set_xlabel("Masse")
ax.set_ylabel("Geschwindigkeit")
```

```python
# Geschwindigkeit und Masse im Vergleich für Zone 2
fig, ax = plt.subplots()
ax.scatter(dataFile2["mass"],dataFile2["velocity"])
ax.set_title("Geschwindigkeit je nach Masse für Zone 2")
ax.set_xlabel("Masse")
ax.set_ylabel("Geschwindigkeit")
```

```python
x1= dataFile1["mass"]
x2=dataFile2["mass"]
y1=dataFile1["velocity"]
y2=dataFile2["velocity"]

plt.scatter(x1,y1,c="blue")
plt.scatter(x2,y2,c="green")
plt.xlabel("Masse")
plt.ylabel("Geschwindigkeit")
plt.title("Geschwindigkeit je nach Masse für Zone kombiniert")
plt.show()
```
