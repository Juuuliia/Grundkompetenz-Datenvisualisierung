```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Daten einlesen
dataFile1 = pd.read_csv("out_1.csv")
dataFile2 = pd.read_csv("out_2.csv")
#Zone hinzufügen
dataFile1["zone"]="1"
dataFile2["zone"]="2"
# Zonen zusammenfügen
steinschlaege = dataFile1.append(dataFile2)
steinschlaege.head(50)
```

```python
steinschlaege.boxplot(column="velocity", by="zone")
plt.title("Geschwindigkeit in m/s nach Ablösungszone")
plt.suptitle(" ")
plt.xlabel("Ablösungszone")
plt.ylabel("Geschwindigkeit in m/s")
plt.show()
```

```python
# Geschwindigkeit und Masse im Vergleich für Zone 1 (Abbildung 1)
fig, ax = plt.subplots()
ax.scatter(dataFile1["mass"],dataFile1["velocity"])
ax.set_title("Geschwindigkeit in m/s je nach Masse für Zone 1")
ax.set_xlabel("Masse")
ax.set_ylabel("Geschwindigkeit in m/s")
```

```python
# Geschwindigkeit und Masse im Vergleich für Zone 2 (Abbildung 2)
fig, ax = plt.subplots()
ax.scatter(dataFile2["mass"],dataFile2["velocity"])
ax.set_title("Geschwindigkeit in m/s je nach Masse für Zone 2")
ax.set_xlabel("Masse")
ax.set_ylabel("Geschwindigkeit in m/s")
```

```python
# Geschwindigkeit der Steine je nach Masse (Abbildung 3)
x1= dataFile1["mass"]
x2=dataFile2["mass"]
y1=dataFile1["velocity"]
y2=dataFile2["velocity"]

plt.scatter(x1,y1,c="blue")
plt.scatter(x2,y2,c="green")
plt.xlabel("Masse")
plt.ylabel("Geschwindigkeit in m/s")
plt.title("Geschwindigkeit in m/s je nach Masse für Zone kombiniert / blau: Zone 1, grün: Zone 2")
plt.show()


# ohne verschiedene Farben
fig, ax = plt.subplots()
ax.scatter(steinschlaege["mass"],steinschlaege["velocity"])
ax.set_title("Geschwindigkeit in m/s je nach Masse für beide Zonen")
ax.set_xlabel("Masse")
ax.set_ylabel("Geschwindigkeit")
```

```python
steinschlaege = steinschlaege.sort_values("date")
steinschlaege.reset_index()
```

```python
# Pie plot welcher Monat Stein gefallen ist

januar = steinschlaege[(steinschlaege["date"]>="2019-01-01") & (steinschlaege["date"]<"2019-02-01")]
anzJanuar = print(januar.shape[0])

februar = steinschlaege[(steinschlaege["date"]>="2019-02-01") & (steinschlaege["date"]<"2019-03-01")]
anzFebruar = print(februar.shape[0])

maerz = steinschlaege[(steinschlaege["date"]>="2019-03-01") & (steinschlaege["date"]<="2019-03-31")]
anzMaerz = print(maerz.shape[0])

steinschlaege[0:37]["monat"] = "Januar"
steinschlaege[38:69]["monat"] = "Februar"
steinschlaege[70:100]["monat"] = "März"
```

```python
# Zeitpunkt der Steinschläge
morgen = (steinschlaege["timestamp"] >= "06:00") & (steinschlaege["timestamp"] < "12:00")
print(sum(morgen))

nachmittag = (steinschlaege["timestamp"] >= "12:00") & (steinschlaege["timestamp"] < "18:00")
print(sum(nachmittag))

abend = (steinschlaege["timestamp"] >="18:00") & (steinschlaege["timestamp"] < "24:00")
print(sum(abend))

nacht = (steinschlaege["timestamp"] <= "24:00") & (steinschlaege["timestamp"] < "06:00")
print(sum(nacht))
```
